# Uppgift 1 d)

"Simulera många punkter slumpmässigt mellan -1 och 1 för x- och y-koordinater,
genom att använda funktionen random.uniform(-1,1)."

```python
import random as rnd

for i in range(100): # upprepar 100 gånger

    x = rnd.uniform(-1,1) #randomizar x koordinat mellan -1 och 1

    y = rnd.uniform(-1,1) #randomizar y koordinat mellan -1 och 1

    print(x,y) #printar randomizad x och y koordinat
```

# Uppgift 1 e) / g)

"Räkna antalet punkter som hamnar inne i cirkeln och dividera med antalet simule-
rade punkter."

"Använd matplotlib och rita punkterna som hamnar inne i cirkeln och punkterna
som hamnar utanför. Använd två olika färger för att särskilja dem åt."

```python
import random as rnd
import matplotlib.pyplot as plt
import math

z = 0

for i in range(10000):
    x = rnd.uniform(-1,1) #randomizar x koordinat mellan -1 och 1
    y = rnd.uniform(-1,1) #randomizar y koordinat mellan -1 och 1
    svar = math.sqrt(x**2+y**2)

    if svar < 1:           
        plt.plot(x,y, '*b')
        z+=1

    else:
        plt.plot(x,y, '*r')

svar = z/10000   

print(svar)
```
