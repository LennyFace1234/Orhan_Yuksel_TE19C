# Uppgift 1 e) / f) / g)

"Räkna antalet punkter som hamnar inne i cirkeln och dividera med antalet simule-
rade punkter."

"Multiplicera resultatet med 4 och skriv ned kommentarer i koden på varför du fick
det resultatet."

"Använd matplotlib och rita punkterna som hamnar inne i cirkeln och punkterna
som hamnar utanför. Använd två olika färger för att särskilja dem åt."

```python
import random as rnd
import matplotlib.pyplot as plt
import math

z = 0 # Ger variabeln z värdet 0

for i in range(10000): 
    x = rnd.uniform(-1,1) #randomizar x koordinat mellan -1 och 1
    y = rnd.uniform(-1,1) #randomizar y koordinat mellan -1 och 1
    svar = math.sqrt(x**2+y**2) # Tar alla simulerade koordinater och räknar ut med hjälp av pythagoras sats vilka som är inne i cirkeln.

    if svar < 1:           
        plt.plot(x,y, '*b')  #Om den är mindre än 1 (Cirkelns radie) så är den innuti cirkeln och visas som blå
        z+=1 # Varje gång en punkt är innuti cirkeln så höjs "z" med 1. z är antalet prickar innuti cirkeln. Den höjs med 1 varje gång det är något innuti cirkeln för att räkna med punkten i de som är innuti cirkeln.

    else:
        plt.plot(x,y, '*r') # Om den är över 1 så är den utanför cirkeln och visas som röd

svar2 = z/10000 # Dividerar antalet punkter som hamnade i cirkeln med antalet simulerade punkter

print(svar2)

print(svar2*4) # Om man gångrar svaret med 4 så ser blir det runt 3.14 vilket är pi. Det kan ha med sättet man räknar ut cirkelns area att göra då pi används för att räkna ut den. 
```

# Uppgift 2 e)

"Designa nu ett program som låter användaren välja start, slut, multipel för ”Burr”
och multipel för ”Birr”. Programmet ska räkna upp antalet ”Burr”, ”Birr” och ”Burr
Birr”"


```python
Startvärde = int(input("Ange startvärde")) # Ber användaren att ange ett startvärde

Slutvärde = int(input("Ange slutvärde")) # Ber användaren att ange ett slutvärde

burr = int(input("Ange tal för burr")) # Ber användaren att ange ett tal för variabeln burr

birr = int(input("Ange tal för birr")) # Ber användaren att ange ett tal för variabeln birr

for i in range(Startvärde,Slutvärde+1): # Detta skriver ut alla värden från startvärdet till ett tal innan slutvärdet. Lägger till + 1 för att den ska skriva upp alla tall från start till slut.
    if i%burr == 0 and i%birr == 0: 
        print("burrbirr", end =" ") # Om talet som skrivs upp är en multipel av variablerna burr och birr så skriver programmet "Burrbirr"

    elif i%burr == 0:
        print("burr", end =" ") # Om talet bara är en multipel av burr så skriver den burr istället för talet

    elif i%birr == 0:
        print("birr", end =" ") # Om talet bara är en multipel av birr så skriver den birr istället för talet

    else:
        print(i, end =" ") # Om talet inte är en multipel av burr, birr eller burr och birr så skrivs bara talet

```

# Uppgift 2 f)

"Ge förslag på hur man kan utveckla programmet vidare och ge det ett försök själv."

```python
# Förslag 1: Man kan göra så att programmet räknar hur många "burr" "birr" och "burrbirr" man får. Programmet kan också räkna den totala antalet av allt tillsammans.

Startvärde = int(input("Ange startvärde"))

Slutvärde = int(input("Ange slutvärde"))

burr = int(input("Ange tal för burr"))

birr = int(input("Ange tal för birr"))

z = 0 # z räknar antalet "burr" som skrivs ut, jag ger den värdet 0

x = 0 # x räknar antalet "birr" som skrivs ut, jag ger den värdet 0

y = 0 # y räknar antalet "burrbirr" som skrivs ut, jag ger den också värdet 0

for i in range(Startvärde,Slutvärde+1):
    if i%burr == 0 and i%birr == 0:
        y+=1 # Varje gång programmet skriver ut "burrbirr" så lägger den till 1 i y. Detta är för att räkna att den skrivit ut burrbirr. Efter den skrivit ut alla burrbirr och lagt till 1 för varje gång burrbirr skrivits ut så kommer y vara värdet av antalet burrbirr.
        print("burrbirr", end =" ")

    elif i%burr == 0:
        z+=1 # Varje gång programmet skriver ut "burr" så lägger den till 1 i z. Detta är för att räkna att den skrivit ut burr. Efter den skrivit ut alla burr och lagt till 1 för varje gång burr skrivits ut så kommer z vara värdet av antalet burr.
        print("burr", end =" ")

    elif i%birr == 0:
        x+=1 # Varje gång programmet skriver ut "birr" så lägger den till 1 i x. Detta är för att räkna att den skrivit ut birr. Efter den skrivit ut alla birr och lagt till 1 för varje gång birr skrivits ut så kommer x vara värdet av antalet birr.
        print("birr", end =" ")

    else:
        print(i, end =" ")

print("Antalet burr:", z,"Antalet birr:", x,"Antalet burrbirr:", y) # Detta gör så att programmet skriver ut z, x och y värdet separat som antalet för burr, birr och burrbirr.
print("Allt tillsammans:", z+y+x,) # Detta adderar z, x och y för att få fram hur många burr, birr och burrbirr vi har tillsammans. Sedan skriver den ut det. 
```
